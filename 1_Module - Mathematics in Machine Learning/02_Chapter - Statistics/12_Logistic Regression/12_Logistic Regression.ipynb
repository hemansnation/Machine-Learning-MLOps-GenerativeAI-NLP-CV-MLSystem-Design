# Binary Classification

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance_data.csv")

df.head()
```

```python
plt.scatter(df.age, df.bought_insurance, marker='+', color='red')
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(df[['age']], df.bought_insurance, train_size=0.9)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(xtrain, ytrain)
```

```python
xtest
```

```python
ypredictions = model.predict(xtest)
ypredictions
```

```python
model.predict_proba(xtest)
```

```python
# binary
0 - 0.5 -> 0
0.5 - 1 -> 1
```

# Digit Dataset

```python
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
digits.data.shape
```

```python
digits.target
```

```python
import numpy as np
```

```python
plt.figure(figsize=(20,4))

for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1,5, index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('%i\n'%label, fontsize=20)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(digits.data, digits.target, train_size=0.75,
                                               random_state=101)
```

```python
model = LogisticRegression()

model.fit(xtrain, ytrain)
```

```python
model.predict(xtest[0].reshape(1,-1))
```

```python
model.predict(xtest[0:10])
```

```python
model.score(xtest, ytest)
```

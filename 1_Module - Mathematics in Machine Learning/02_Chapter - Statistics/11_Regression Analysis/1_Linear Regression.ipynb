```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("homeprices.csv")

df.head()
```

```python
plt.xlabel("Area")
plt.ylabel("Price")

plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
area = df[['area']]
area
```

```python
price = df.price

price
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model
```

```python
model.fit(area, price)
```

```python
# m = slope = gradient = 135.78767123
model.coef_
```

```python
# c = intercept = 180616.43835616432
model.intercept_
```

```python
y = 135.78(x) + 180616.43
```

```python
x = 3100
```

```python
135.78*3100 + 180616.43
```

```python
model.predict([[3100]])
```
